```python
import io
import pandas as pd

#Reading CSV File
df = pd.read_csv("AAPL.csv")
#Getting the Date column which contains Day, Month, Year
df['Month'] = pd.to_datetime(df["Date"])
df['Year'] = pd.to_datetime(df["Date"])
total_prices = {}
#Loop Through 2019-2023
for year in (2019,2020,2021,2022,2023):
#August Data from 2019 - 2023
  august_data = df[(df['Month'].dt.month == 8) & (df['Year'].dt.year == year)]
  total_prices[year] = august_data['Close'].mean()
#Printing the average price from August of that year
print(total_prices)


```
